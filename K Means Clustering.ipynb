```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np 
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features= 2, centers= 4, cluster_std=1.8)
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

```python
from sklearn.cluster import KMeans
KM = KMeans(n_clusters=16)
KM.fit(data[0])
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize = (11,7))
ax1.scatter(data[0][:,0],data[0][:,1], c = KM.labels_)
ax2.scatter(data[0][:,0],data[0][:,1], c = data[1])
```
